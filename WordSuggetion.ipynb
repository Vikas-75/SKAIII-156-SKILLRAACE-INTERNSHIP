```python
from ipywidgets import interact, widgets
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load pre-trained model and tokenizer
model_name = 'gpt2'
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

def predict_next_words(text, num_words=5):
    # Encode the input text
    input_ids = tokenizer.encode(text, return_tensors='pt')

    # Generate multiple sequences with the model
    output_sequences = model.generate(
        input_ids,
        max_length=len(input_ids[0]) + 10,
        num_return_sequences=num_words,
        no_repeat_ngram_size=2,
        top_k=50,
        top_p=0.95,
        temperature=1.0,
        do_sample=True,
        pad_token_id=model.config.pad_token_id,
        eos_token_id=model.config.eos_token_id,
        early_stopping=True
    )

    # Decode the output sequences
    predicted_texts = [tokenizer.decode(seq, skip_special_tokens=True) for seq in output_sequences]

    return predicted_texts

# Function to handle text input and show suggestions
def handle_text_input(text):
    if text:
        predicted_words = predict_next_words(text, num_words=5)
        suggestion_output.clear_output()
        with suggestion_output:
            for i, word in enumerate(predicted_words, start=1):
                print(f"{i}. {word}")

# Create interactive text input and suggestion output
text_input = widgets.Text(
    value='',
    placeholder='Type here...',
    description='Input:',
    continuous_update=True
)

suggestion_output = widgets.Output()

# Interact function to display text input and handle suggestions
interact(handle_text_input, text=text_input)

# Display output for suggestions
display(suggestion_output)
```
