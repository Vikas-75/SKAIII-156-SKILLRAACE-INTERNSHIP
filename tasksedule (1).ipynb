```python
from ipywidgets import interact, widgets, Layout, Button
from datetime import datetime
import pytz  # For timezone support
# List to store tasks
tasks = []

# Timezone (example: New York timezone)
tz = pytz.timezone('America/New_York')
# Function to add a task
def add_task(description, due_date):
    due_date = tz.localize(datetime.combine(due_date, datetime.min.time()))  # Localize due date
    tasks.append({"description": description, "due_date": due_date})
    print(f"Task added: '{description}' due on {due_date.strftime('%Y-%m-%d %H:%M %Z')}")

# Function to view all tasks
def view_tasks():
    if tasks:
        print("\nYour tasks:")
        for i, task in enumerate(tasks, start=1):
            print(f"{i}. {task['description']} - Due on {task['due_date'].strftime('%Y-%m-%d %H:%M %Z')}")
    else:
        print("You have no tasks scheduled.")
# Function to add a task
# Function to add a task
def add_task(description, due_date):
    if due_date is None:
        print("Please select a due date.")
        return  # Exit function if due_date is None

    due_date = tz.localize(datetime.combine(due_date, datetime.min.time()))  # Localize due date
    tasks.append({"description": description, "due_date": due_date})
    print(f"Task added: '{description}' due on {due_date.strftime('%Y-%m-%d %H:%M %Z')}")


# Function to view all tasks
def view_tasks():
    if tasks:
        print("\nYour tasks:")
        for i, task in enumerate(tasks, start=1):
            print(f"{i}. {task['description']} - Due on {task['due_date'].strftime('%Y-%m-%d %H:%M %Z')}")
    else:
        print("You have no tasks scheduled.")
# Text input widget for task description
description_input = widgets.Text(
    description='Task Description:',
    layout=Layout(width='50%')
)

# Date picker widget for task due date
due_date_input = widgets.DatePicker(
    description='Due Date:',
    layout=Layout(width='50%')
)
# Function to handle user input and add tasks
def handle_user_input(description, due_date):
    add_task(description, due_date)

# Interact function to link user input with task management
interact(handle_user_input, description=description_input, due_date=due_date_input)
# Function to handle click event of the View Tasks button
def on_view_tasks_button_clicked(button):
    view_tasks()

# Create a button widget for viewing tasks
view_tasks_button = Button(description="View Tasks")
view_tasks_button.on_click(on_view_tasks_button_clicked)

# Display the button
display(view_tasks_button)
```
